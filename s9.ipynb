```python
import numpy as np  
import matplotlib.pyplot as plt 

# Generate 1000 random numbers distributed uniformly between [0, 1]
random_numbers = np.random.rand(1000)

# Create a histogram with 100 bins
plt.hist(random_numbers, bins=100, edgecolor='black')

# Labeling axes 
plt.xlabel('Random Numbers')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.title('Uniformly Distributed Random Numbers Histogram')


plt.savefig('random_numbers_histogram.pdf')

# Display the plot for immediate visualization
plt.show()
```
