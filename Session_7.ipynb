```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#Creating an array x ranging from [0, 1] in 100 steps
x = np.linspace(0, 1, 100)

#Defining a function that returns exp(x)
def exponential_function(x):
    return np.exp(x)

#Using the function to set y=exp(x)
y = exponential_function(x)

#Plotting x vs. y
plt.plot(x, y, label='exp(x)')

# Adding labels to the axes
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')

# Adding a title to the plot
plt.title('Awesomeness over Time')

# Adding a legend to the plot
plt.legend()

#Saving the figure as a PDF
plt.savefig('awesomeness_plot.pdf')

# Displaying the plot
plt.show()
```
