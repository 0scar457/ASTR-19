```python
import numpy as np
import matplotlib.pyplot as plt

#Create an array x ranging from [0, 1] in 100 steps
x = np.linspace(0, 1, 100)

#Define functions for sin(x) and cos(x)
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)

#Create a multipanel plot (1 row, 2 columns)
#lot sin(x) vs. x in the left panel and cos(x) vs. x in the right panel
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

#Plot sin(x) vs. x
ax[0].plot(x, sin_function(x), label='sin(x)', color='blue')
ax[0].set_title('sin(x) vs. x')  # Set the title
ax[0].set_xlabel('x')  # Set the x-axis label
ax[0].set_ylabel('sin(x)')  # Set the y-axis label
ax[0].legend()  # Display legend

#olot cos(x) vs. x
ax[1].plot(x, cos_function(x), label='cos(x)', color='red')
ax[1].set_title('cos(x) vs. x')  # Set the title
ax[1].set_xlabel('x')  # Set the x-axis label
ax[1].set_ylabel('cos(x)')  # Set the y-axis label
ax[1].legend()  # Display legend

#adjust layout to prevent overlap
plt.tight_layout()

#Save the figure as a PDF
plt.savefig('sin_cps_plot.pdf')

plt.show()
```
