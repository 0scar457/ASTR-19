```python
import numpy as np
import matplotlib.pyplot as plt

#normal distribution 
mu, sigma = 0, 1  # mean and standard deviation
random_numbers = np.random.normal(mu, sigma, 1000)

#the histogram
plt.hist(random_numbers, bins=100, density=True, alpha=0.7, color='blue', edgecolor='black')

# Labeling axes and title
plt.title('Histogram of Random Numbers (Normal Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.savefig('random_numbers_histogram.pdf')

plt.show()
```
